```python
# ---------------
# Data Cleaning
# ---------------
```

```python
df=df.drop_duplicates()
```

```python
df=df[df['Quantity']>0]
df=df[df['UnitPrice']>0]
```

```python
# ----------------------
# Data Type Conversion
# ----------------------
```

```python
df['CustomerID']=df['CustomerID'].astype(str)
```

```python
df['InvoiceDate']=pd.to_datetime(df['InvoiceDate'])
```

```python
# ---------------------
# Feature Engineering
# ---------------------
```

```python
df['TotalPrice']=df['Quantity']*df['UnitPrice']
```
